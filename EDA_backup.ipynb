```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df[["math score","reading score","writing score"]].hist(figsize=(10,5), edgecolor='black')
plt.suptitle("Distribution of Student Scores")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df[["math score","reading score","writing score"]])
plt.title("Score Comparison (Math vs Reading vs Writing)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="gender", y="math score", data=df)
plt.title("Math Score by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="lunch", y="math score", data=df)
plt.title("Effect of Lunch Type on Math Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="test preparation course", y="math score", data=df)
plt.title("Test Preparation vs Math Score")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[["math score","reading score","writing score"]])
plt.show()
```
